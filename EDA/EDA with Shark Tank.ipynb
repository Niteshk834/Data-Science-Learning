# Q14. Find the equity percent that each sharks gets? 

```python
df.columns
```

```python
ashneer=df[df["ashneer_deal"]==1]["equity_per_shark"].sum()
anupam=df[df["anupam_deal"]==1]["equity_per_shark"].sum()
aman=df[df["aman_deal"]==1]["equity_per_shark"].sum()
namita=df[df["namita_deal"]==1]["equity_per_shark"].sum()
vineeta=df[df["vineeta_deal"]==1]["equity_per_shark"].sum()
peyush=df[df["peyush_deal"]==1]["equity_per_shark"].sum()
ghazal=df[df["ghazal_deal"]==1]["equity_per_shark"].sum()
```

```python
equity_data={"ashneer": ashneer,"anupam": anupam,"aman":aman,"namita":namita,"vineeta":vineeta,"peyush":peyush,"ghazal":ghazal}
```

```python
df13=pd.DataFrame(data=equity_data.items(),columns=["Shark_name","equity %"])
```

```python
df13
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df13,x="Shark_name",y="equity %")
plt.ylabel("equity % that each shark gets")
plt.show()
```
