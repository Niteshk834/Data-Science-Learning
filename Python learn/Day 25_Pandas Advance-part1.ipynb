```python
# Let's unserstand about merge and joint operation
```

```python
data1=pd.DataFrame({"key1":(2,3,4,7),"key2":(4,3,5,6),"key3":(2,6,3,5)})

data2=pd.DataFrame({"key1":(22,3,4,7),"key4":(40,43,65,16),"key5":(2,16,23,45)})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2) # result is based on the common key between both data
```

```python
pd.merge(data1,data2,how="left") # left means data1 are as it is and after that match value of key1 in both
                                 # data and matched data in which row ,the value of key4 and key5 we got.
```

```python
pd.merge(data1,data2,how="right") # right means data2 are as it is and after that match value of key1 in both
                                 # data and matched data in which row ,the value of key2 and key3 we got.
```

```python
pd.merge(data1,data2,how="outer") # we got all the left and right value and matched that value of key1
                                  # if value in key match in left then as it is value of key2 and key3 we
# write and value matched in right too then key4 and key5 value as it is we write. if not matched value 
# will be NaN and can present either left and right.
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,how="cross") # basically it take first row of data1 and combined with all row of data2
```

```python
data1
```

```python
data2
```

```python
# if we want purposly merge on particular column
pd.merge(data1,data2,how="left",left_on="key2",right_on="key4")
```

```python
pd.merge(data1,data2,how="left",left_on="key3",right_on="key5")
```

```python
import pandas as pd
```

```python
data1=pd.DataFrame({"key1":(2,3,4,7,5),"key2":(4,2,3,5,6),"key3":(2,6,4,3,5)},index=["a","b","c","d","e"])

data2=pd.DataFrame({"key4":(2,3,4,7,1),"key5":(4,6,5,2,3),"key6":(2,1,3,5,4)},index=["a","b","f","g","h"])
```

```python
data1
```

```python
data2
```

```python
data1.join(data2) # column not overlap or same name in both data for join operation perform.

# In case of join operation it matches basically the indexes not the number

# Output give left join, means that data1 are as it is.
```

```python
data1.join(data2,how="inner")
```

```python
data1.join(data2,how="outer")
```

```python
data1.join(data2,how="right")
```

```python
data1.join(data2,how="left")
```

```python
data1.join(data2,how="cross")
```
