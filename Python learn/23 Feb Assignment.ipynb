# Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?



## Answer:-

### Some of the most common functions we can use to manipulate data in pandas DataFrame are following:

#### 1. head() and tail(): 

These functions are used to show the first or last n rows(By default- 5 rows) of a DataFrame, respectively.They are useful for quickly getting a sense of the structure and content of a DataFrame.

#### 2. describe():

This function provides summary statistics for the numerical columns in a DataFrame, such as the count, mean, standard deviation, minimum, and maximum. It can be useful for getting a quick overview of the distribution of data in a DataFrame.

#### 3. sort_values(): 

This function is used to sort a DataFrame by one or more columns. 

For example-  We might use sort_values() to sort a DataFrame of customer orders by order date, so that we can see which orders were placed first.

#### 4. groupby():

This function is used to group rows in a DataFrame by one or more columns, and then perform aggregate functions on the groups.

For example-  we might use groupby() to group a DataFrame of sales data by product category, and then calculate the total sales and average price for each category.

#### 5. drop(): 

This function is used to drop rows or columns from a DataFrame. 

For example- weou might use drop() to remove rows with missing values, or columns that are not relevant to our analysis.

#### 6. fillna():

This function is used to fill missing values in a DataFrame.

For example- we might use fillna() to replace missing values with the mean or median value of the column.

#### 7. apply():

This function is used to apply a function to one or more columns in a DataFrame. 

For example- we might use apply() to apply a function that converts all values in a column to lowercase.

#### 8. merge(): 

This function is used to combine two or more DataFrames based on a common column.

For example- we might use merge() to combine a DataFrame of customer orders with a DataFrame of customer information, based on a common customer ID.

```python
# An example of one of these function like head and tail is :
```

# Q6. Which of the following is mutable in nature Series, DataFrame, Panel?



## Answer:-

In Pandas, both Series and DataFrame are mutable in nature, meaning their values can be modified after creation. 

However, Panel objects are not considered mutable and are generally recommended to be avoided in favor of using multi-dimensional NumPy arrays for higher-dimensional data. Panel was a three-dimensional data structure in pandas that has been deprecated and removed in version 1.0.0.

Also one thing is that, while both Series and DataFrame are mutable, it is generally recommended to avoid modifying them in place and instead create new copies with the desired changes.

```python
# Example of mutable of dataseries
```

```python
import pandas as pd
data=pd.Series([1,2,3,4,5])
print(data)
```

```python
data[3]=20 # Here at the 3rd index postion starting from 0, data will be change.
print(data)
```

```python
# Example of mutable of dataframe
```

```python
import pandas as pd

df=pd.DataFrame({"Name":["Alice","Bob","Claire"],
      "Age":[25,30,27],
      "Gender":["Female","Male","Female"]})
# print the data frame
print(df)
```

```python
# we want to add new row in data frame
df.loc[3]=["Nitesh",32,"Male"]
print(df)
```

# Q7. Create a DataFrame using multiple Series. Explain with an example

## Answer:-

```python
import pandas as pd
# create multiple series
vehicle=pd.Series(["car","Bus","Truck","tractor","HAM","Bike","Auto"])
vehicle_class=pd.Series(["2 Axle","3 Axle","4 Axle","2 Axle",">7 Axle","No Axle","1 Axle"])
Journey_Type=pd.Series(["Return Journey","Single Journey","Single Journey","Single Journey",
                          "Single Journey","Single Journey","Return Journey"])
Fare=pd.Series([145,160,210,150,560,0,0])
speed=pd.Series([80,60,55,30,10,60,40])
Owner=pd.Series(["Nitesh","Rahul","Rambabu","kumar","nagarjun","Sachin","Ishaq Khan"])

# Creating a DataFrame by combining the Series
df=pd.DataFrame({"vehicle": vehicle,"vehicle_class":vehicle_class,"Journey_Type":Journey_Type,
                "Fare": Fare,"speed":speed,"Owner":Owner})
# Print or displaying the DataFrame
print(df)
```
