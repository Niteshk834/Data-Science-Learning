# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.



## Answer:-

```python
df
```

### Now create a function

```python
import pandas as pd
# define function
def reindex_df(df):
    new_index = pd.RangeIndex(start=1, stop=2*len(df)+1,step=2)
    # In stop=2*len(df)+1 we used to multiply by 2 of length of origninal index becuase after new indexing 
    # index will be in range if step size =2
    new_df = df.set_index(new_index)
    return new_df    
```

Here's the basic syntax of the read_excel() function:

```python
pd.read_excel(io, sheet_name=0, header=0, index_col=None, ...)
```

Here are the arguments of the read_excel() function:

### io/ file_path: 
This argument specifies the path of the Excel file to be read. It can be a string that represents the file path or a file-like object such as a StringIO object.

### sheet_name:
This argument specifies which sheet of the Excel file to read. By default, it reads the first sheet. You can pass the sheet name as a string or the sheet index as an integer. You can also pass a list of sheet names or indices to read multiple sheets.

### header:
This argument specifies which row in the Excel file should be used as the column labels for the DataFrame. By default, it uses the first row. You can pass an integer to specify the row number or a list of integers to specify multiple rows.
### index_col:
This argument specifies which column in the Excel file should be used as the index for the DataFrame. By default, it uses a range of integers starting from 0. You can pass an integer to specify the column number or a string to specify the column name.

# Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.



## Answer:-

```python
# create data frame
df=pd.DataFrame({"Values":[1,2,3,4,5,6,7,8]})
```

```python
df["Values"]
```

## Now create a function

```python
import pandas as pd
import numpy as np # NumPy arrays provide a fast and efficient way to perform mathematical operations
                   # on arrays of data. So our moto is to firstly stored values of dataframe in Numpy array
def calculate_values(df):
    values=df["Values"] # The values of the "Values" column are extracted from the DataFrame 
                        # and stored in a NumPy array
    mean=np.mean(values)
    median=np.median(values)
    std=np.std(values)
    return {"mean":mean,"median":median,"std":std}   
```

```python
# call the function
calculate_values(df)
```

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

## Answer:-

```python
import pandas as pd
import numpy as np
# Create a list of random sales data
np.random.seed(40) # for seed every execution we get the same number
sales_data=np.random.randint(500,1000,size=15)
# Create a list of dates
dates=pd.date_range(start="2023-01-01",end="2023-01-15",freq="D") # D use for daily frequency
df=pd.DataFrame({"Sales":sales_data,"Date":dates})
```

```python
df
```

### Now create a function

```python
def add_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df
```

```python
# call the function
add_moving_average(df)
```

The function should return the modified DataFrame.


## Answer:-

```python
# create data
dates=pd.date_range(start="2023-01-01",end=" 2023-01-05")
# create frame
df=pd.DataFrame({"Date":dates})
```

```python
df
```

```python
# extract the weekdays from data frame
df["Date"].dt.day_name()

 # .dt is a pandas property that provides access to datetime-related properties of a column of pandas
 # datetime objects.

# .day_name() is a method that is used to extract the name of the day of the week from each timestamp 
# in the datetime column. This method returns the name of the day of the week as a string 
# (e.g. "Monday", "Tuesday", etc.) for each value in the column.
```
