Note: Depending on the specific requirements of the analysis, the steps mentioned above are not necessarily sequential and can be performed in a different order.

# Q5. What is meant by standard deviation and variance for a numerical data?


### Answer:-

Standard deviation and variance are both measures of the spread or dispersion of a set of numerical data.

Variance is the average of the squared differences from the mean of a set of data. It gives a measure of how far the data points are from the average. The formula for variance is:

variance = (sum of (data point - mean)^2) / (number of data points)


```python
df3
```

```python
# calculate mean
age_mean=df3["Age"].mean()
```

```python
age_mean
```

```python
# calculate the variance of the age column
age_var = sum((i - age_mean) ** 2 for i in df3["Age"]) / (len(df3["Age"]) - 1)
```

```python
age_var
```

This means that the ages of the passengers on the Titanic are somewhat spread out from the mean age.

Standard deviation is the square root of variance. It is a measure of how spread out the data is relative to the mean. The formula for standard deviation is:

```python
# calculate the standard deviation of the age column
age_std=age_var**0.5
```

```python
age_std
```

Both variance and standard deviation are commonly used in statistics to describe the variability or dispersion of a set of data. A higher variance or standard deviation indicates a greater spread of the data points, while a lower variance or standard deviation indicates a more tightly clustered set of data points around the mean.

# Q6. What is meant by imbalance in the categorical data and how do we handle that?


### Answer:-

Imbalance refers to the situation where one or more categories have significantly fewer observations than the others. This can be a problem because it can bias the results of data analysis or modeling towards the overrepresented categories, making it difficult to draw accurate conclusions or predictions about the underrepresented categories.

#### There are several methods to handle imbalanced categorical data:

Suppose we have a dataset of credit card transactions where the majority of the transactions are legitimate and only a small portion are fraudulent. In this case, we have an imbalanced dataset where the minority class is the fraudulent transactions.

Let's say our goal is to build a machine learning model to predict whether a transaction is fraudulent or not. However, since the dataset is imbalanced, our model might not perform well on the minority class (i.e., the fraudulent transactions) and might be biased towards the majority class (i.e., legitimate transactions).

To address this issue, we can use one or more of the following strategies:

### 1. Resampling:

#### Oversampling:

We can increase the number of fraudulent transactions by generating synthetic samples of the minority class using techniques like SMOTE (Synthetic Minority Over-sampling Technique). This involves creating new instances of the minority class by interpolating between existing samples of that class. For instance, we could generate new synthetic fraudulent transactions by perturbing the features of existing fraudulent transactions.

#### Undersampling: 

We can reduce the number of legitimate transactions by randomly selecting a subset of the majority class to balance the dataset. However, this method may result in a loss of information and can lead to overfitting.

### 2. Class weighting:

We can assign higher weights to the minority class to increase the importance of the fraudulent transactions during model training. For instance, we could assign a weight of 10 to the fraudulent transactions and a weight of 1 to the legitimate transactions.

### 3. Synthetic data generation:

We can generate synthetic data points for the minority class using techniques like data augmentation or SMOTE. For instance, we could create new fraudulent transactions by randomly modifying the features of existing fraudulent transactions.

### 4. Algorithm selection:

We can choose an algorithm that is less sensitive to class imbalance. For instance, decision trees and random forests can handle imbalance well, while logistic regression and SVM may require rebalancing.

### 5. Evaluation metrics:

We should use evaluation metrics that are more appropriate for imbalanced datasets, such as precision, recall, F1-score or AUC-ROC. For instance, we could calculate the precision of our model as the number of true positives divided by the sum of true positives and false positives. This would give us a more accurate measure of how well our model is performing on the minority class.

# Q8. Explain IQR method to remove outliers in the numerical data?

### Answer:-

The IQR (interquartile range) method is a robust method to remove outliers from numerical data. It involves calculating the difference between the third quartile (Q3) and the first quartile (Q1) of the dataset. This difference is referred to as the IQR.

Here are the steps to apply the IQR method to remove outliers:

Calculate the first quartile (Q1) and the third quartile (Q3) of the dataset. The quartiles divide the dataset into four equal parts, with Q1 being the value at the 25th percentile and Q3 being the value at the 75th percentile.

Calculate the IQR by subtracting Q1 from Q3: IQR = Q3 - Q1.

Identify the lower and upper bounds for outliers by computing the following:

Lower bound = Q1 - 1.5 * IQR
Upper bound = Q3 + 1.5 * IQR
Remove any values in the dataset that fall below the lower bound or above the upper bound. These values are considered outliers.
Note that the value 1.5 is a commonly used threshold for the number of standard deviations from the mean, but it can be adjusted depending on the specific dataset and the desired level of outlier removal.

It's important to keep in mind that while the IQR method is a useful tool for identifying and removing outliers, it may not be appropriate for all datasets or situations. Additionally, removing outliers can have a significant impact on the distribution and analysis of the data, so it should be done thoughtfully and with a clear understanding of the implications.

#### Let's understand with Example

```python
# Calculate the first quartile (Q1) and the third quartile (Q3) of the age column.
```

```python
import numpy as np
```

```python
# Let's assume that in age column of titanic dataset have follwing value:
age = np.array([22, 38, 26, 35, 100, 35, 28, 54, 2, 27, 14, 4, 58, 20, 39, 14, 55, 2, 31, 35])
            
```

```python
# calculate Q1
Q1=np.percentile(age,25)
```

```python
Q1
```

```python
# calculate Q3
Q3=np.percentile(age,75)
```

```python
Q3
```

```python
# calculate IQR
IQR=Q3-Q1
```

```python
IQR
```

```python
# Identify Lower bound
lower_bound=Q1-1.5*IQR
```

```python
lower_bound
```

```python
# Identify upper bound
upper_bound=Q3+1.5*IQR
```

```python
upper_bound
```

```python
# Remove any value in age column that fall below lower bound and above upper bound
age_without_outlier=age[(age>=lower_bound)&(age<=upper_bound)]
```
