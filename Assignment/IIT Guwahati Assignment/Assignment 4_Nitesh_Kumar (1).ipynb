# Q1. Why Do We Need Data Visualization? Explain with Examples.

## Answer:-

Data visualization is a powerful tool for making sense of complex information and data. It allows us to present data in a visual format that is easy to understand and interpret, making it more accessible to a wider audience.

Here are some reasons why data visualization is important and some examples of how it can be used:

#### 1. To uncover insights and patterns:

Data visualization can help us identify patterns and trends that might not be apparent from raw data. 

##### For example-
A line graph can show how a company's sales have changed over time, revealing seasonal patterns or changes in consumer behavior.

#### 2. To communicate information effectively: 

Visualizations can make complex information more accessible and easier to understand. 

##### For example-

A map can show the spread of a disease or the distribution of a population, while a bar chart can compare the performance of different products or services.

#### 3. To support decision-making:

Data visualization can help decision-makers to quickly identify key trends or patterns, allowing them to make more informed decisions.

##### For example-
A dashboard might show real-time data on website traffic or customer satisfaction scores, enabling managers to respond quickly to changes in performance.

#### 4. To tell a story: 

Visualizations can be used to tell a story and make data more engaging.

##### For example-
An infographic might use a combination of charts, maps, and images to tell a story about a social issue or environmental problem.

```python
flights=sns.load_dataset("flights") # load predefined dataset set flights
```

```python
flights
```

```python
# check columns
flights.columns
```

The sns.heatmap function requires a two-dimensional array of data to create a heatmap, but the flights dataset is in long format with three variables: year, month, and passengers.

To create a heatmap from the flights dataset, we first need to pivot the data so that the rows correspond to month, the columns correspond to year, and the values correspond to the number of passengers.

```python
# Reshape the data into a pivot table
flight_data=flights.pivot("month","year","passengers")
```

If we want to remove the FutureWarning, we can use keyword arguments instead of positional arguments when calling the pivot

```python
flight_data = flights.pivot(index="month", columns="year", values="passengers")
# Now no future warning after execution
```

```python
# plot heatmap
sns.heatmap(data=flight_data,annot=True,fmt="d") 

# fmt="d" parameter formats the numerical values as integers.
# annot=True parameters adds the numerical value of each cell to the heatmap.
plt.show()
```
