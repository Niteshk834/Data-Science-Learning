##### Reason of SMOTE(Synthetic Minority oversampling Technique) Use

In upsampling, we increasing same datapoints again amd again. Means more data points add in the same place. This technique is not as good because my data still is
in same position means that variance not increse at all.
In order to increase the varience we use interpolating technique.

In SMOTE we try to join two nearest point and add the data point b/w these join line.

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],
                    random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=["f1","f2"])
df2=pd.DataFrame(y,columns=["target"])
```

```python
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df
```

```python
final_df["target"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df["f1"],final_df["f2"],c=final_df["target"])
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# Now transform the dataset
oversample=SMOTE()
```

```python
X,y=oversample.fit_resample(final_df[["f1","f2"]],final_df["target"])
```
